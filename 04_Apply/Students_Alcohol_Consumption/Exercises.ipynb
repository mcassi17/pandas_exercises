### Step 9. Create a function called majority that return a boolean value to a new column called legal_drinker

```python
def majority(row):
    val = None
    if row >= 21:
        val = True
    else:
        val = False
    return val
```

```python
filtered['legal_drinker'] = filtered['age'].apply(majority)
filtered.head()
```
